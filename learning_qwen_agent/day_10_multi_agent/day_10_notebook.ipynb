## Part 3: Human-in-the-Loop

### Adding Human Participants

From `group_chat_demo.py`, you can add humans with `is_human: True`:

```python
agents = [
    {
        'name': 'User',
        'description': 'A human team member',
        'is_human': True  # This marks it as human!
    },
    {
        'name': 'Assistant',
        'description': 'AI helper',

    }

```

When `is_human: True`, the GroupChat will wait for human input!

````python
# Example with human participant
team_with_human = {
    'background': 'Team discussion',
    'agents': [
        {
            'name': 'Alice',
            'description': 'Team lead (human)',
            'is_human': True
        },
        {
            'name': 'CodeBot',
            'description': 'AI coding assistant',
            'instructions': 'You help with coding questions. Be helpful.'
        },
        {
            'name': 'DocBot',
            'description': 'AI documentation assistant',
            'instructions': 'You help write documentation.'
        }
    ]
}

print("âœ… Team with human created")
for agent in team_with_human['agents']:
    role = "(Human)" if agent.get('is_human') else "(AI)"
    print(f"  - {agent['name']} {role}: {agent['description']}")
````

## Part 4: @Mention System

```python
user_message = "@CodeBot can you write a factorial function?"
```

The system:
1. Detects @CodeBot
2. Routes message to CodeBot
3. CodeBot responds
4. Conversation continues

GroupChat tracks mentioned agents:
```python

    ...
```

## Part 6: Agent Configuration Fields

### Complete Agent Schema

From the official examples:

```python
{
    'name': str,                    # Required: Agent name
    'description': str,             # Required: What agent does
    'is_human': bool,               # Optional: Is this a human?
    'instructions': str,            # Optional: System message
    'knowledge_files': List[str],   # Optional: Files for RAG
    'selected_tools': List[str]     # Optional: Tools this agent can use
}
```

**Key Points:**
- `is_human: True` â†’ Wait for human input
- `instructions` â†’ Agent's system message
- `selected_tools` â†’ Each agent can have different tools!
- `knowledge_files` â†’ Per-agent RAG

````python
# Example: Agents with different capabilities
specialized_team = {
    'background': 'Technical support team',
    'agents': [
        {
            'name': 'Customer',
            'description': 'Customer with a question',
            'is_human': True
        },
        {
            'name': 'CodeHelper',
            'description': 'Helps with code',
            'instructions': 'You help debug code. Use code_interpreter.',
            'selected_tools': ['code_interpreter']
        },
        {
            'name': 'DocHelper',
            'description': 'Helps find documentation',
            'instructions': 'You search documentation and explain APIs.',
            'knowledge_files': [],  # Could add docs here
            'selected_tools': []
        }
    ]
}

print("Specialized team:")
for agent in specialized_team['agents']:
    tools = agent.get('selected_tools', [])
    tool_str = f" [Tools: {', '.join(tools)}]" if tools else ""
    print(f"  - {agent['name']}{tool_str}")
    print(f"    {agent['description']}")
````

## Part 7: How GroupChat Works

### The Conversation Flow

1. **User sends message** (or agent starts)
2. **GroupChat decides** who should respond
3. **Selected agent** generates response
4. **Response added** to conversation
5. **Repeat** until done

### Decision Logic:
- If @mentioned: Route to that agent
- If is_human needed: Wait for input
- Otherwise: Auto-select based on context

### From group_chat_demo.py:
```python
for response in bot.run(messages, need_batch_response=False, mentioned_agents_name=[]):
    # Process responses
    pass
```

## Part 8: Coordination Patterns

### Pattern 1: Sequential
Agents take turns in order:
1. Planner â†’ Creates plan
2. Coder â†’ Writes code
3. Tester â†’ Tests code
4. Reviewer â†’ Reviews all

### Pattern 2: Router
One agent routes to specialists:
- User question â†’ Router
- Router decides: Code/Doc/Support
- Specialist handles it

### Pattern 3: Collaborative
Agents discuss freely:
- Multiple agents can contribute
- @mentions to call specific agents
- Human can intervene anytime

GroupChat supports all patterns!
