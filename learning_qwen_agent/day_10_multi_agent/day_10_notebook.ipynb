## GroupChat Agent

GroupChat manages multiple agents:
```python
from qwen_agent.agents import GroupChat

cfg = {
    'background': 'Team discussion',
    'agents': [
        {'name': 'Coder', 'instructions': 'Write code'},
        {'name': 'Reviewer', 'instructions': 'Review code'}
    ]
}

bot = GroupChat(agents=cfg, llm=llm_cfg)
```
